# Using PyPI with pip install

PyPI is a repository for open source third party Python packages.
If you've worked of other programming languages it's similar to:

    Ruby - gen's in the Ruby world 
    PHP - Packagist
    perl - CPAN
    node.js - npm
    python - PyPI

So far we've only used libraries that come internally of python and these are known as the **standard library or the built in libraries for Python** things like the math library.

Now there are many other libraries available that people have open source and shared on PyPI and these are also called **packages** and we can use pip installed at the command line to install these **external packages**.


# Find Packages

Find more packages at https://pypi.org/.


# PIP

PIP is a package manager for Python packages, or modules if you like.


**Pip** is a simple way to download packages at your command line directly from the PyPI repository.


By installing Python from Python.org or by installing the Anaconda's distribution as shown earlier in this course, you actually have already installed Pip. 

Note: If you have Python version 3.4 or later, PIP is included by default.



Now there are packages already created for almost any use case you can think of.


Python is a hugely popular language and it's growing very quickly.

FOR WEB DEVELOPMENT : Django and flask packages are available

#List Packages

Use the list command to list all the packages installed on your system:

    C:\Users\Your Name\AppData\Local\Programs\Python\Python36-32\Scripts>pip list




# Remove a Package

Use the uninstall command to remove a package:


    Uninstall the package named "camelcase":

    C:\Users\Your Name\AppData\Local\Programs\Python\Python36-32\Scripts>pip uninstall camelcase


Uninstalling camelcase-02.1:

  Would remove:

    c:\users\Your Name\appdata\local\programs\python\python36-32\lib\site-packages\camecase-0.2-py3.6.egg-info
    c:\users\Your Name\appdata\local\programs\python\python36-32\lib\site-packages\camecase\*
    Proceed (y/n)?


Press y and the package will be removed.
