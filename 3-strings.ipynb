# Slicing

To get the substring from the string.


    Basic Format of Slicing:
    
    string_variable[starting_index : stop_index(not including the index value) : step_size]


First column represent the characters in the string

Second column represent the corresponding index value.

    H 0
    e 1
    l 2
    l 3
    o 4
      5
    W 6
    o 7
    r 8
    l 9
    d 10


```python
# Get the whole string as a substring
# Alternative way - 3rd way to print the string (in order to understand the concept only)

print(s[:])
```

```python
#Get the substring from the first index position to last index position

print( s[1:] )

#Get the substring from the second index position to last index position

print( s[1:] )

#Get the substring from the fifth index position to last index position

print( s[5:] )

#Get the word at the last index position

print( s[10:] )
```

```python
# Grab everything upto the 3rd index position
print(s[:3])

# Grab everything upto the 3rd index position
print(s[:4])

# Grab everything upto the 5th index position
print(s[:5])

# Grab everything upto the last index position
# 4th way to print the whole string ( in order to understand the concept)
print(s[:10])
```

**Note the above slicing. Here we're telling Python to grab everything from 0 up to 3. It doesn't include the 3rd index.** 

You'll notice this a lot in Python, where statements and are usually in the context of 'up to, but not including'.

```python
# Note that there is no change to the original s. All the changes are made in the copy of the string.
s
```

We can also use negative indexing to go backwards. Similarly we can also use the negative indexing ( to read the srting in reverse order) during slicing.

```python
print(s[-1])
```

```python
print(s[-1:])
```

```python
print(s[-2:])
```

```python
# Grab everything but not the last two letters. Notice the negative value at the stop_index position
s[:-2]
```

**We can also use index and slice notation to grab elements of a sequence by a specified step size (the default is 1). For instance we can use two colons in a row and then a number specifying the frequency to grab elements.** For example:

```python
# Grab everything, but go in steps size of 1
# Yeahh!! Another way just in order to understand the concept.

s[::1]
```

```python
# Grab everything, but go in step sizes of 3
s[::3]
```

```python
# We can use this to print a string backwards
s[::-1]
```

```python
# We can use this to print a string backwards by the step 2 in reverse order
s[::-2]
```
